```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("maaslar.csv")

xdf = df.iloc[:,1:2]
ydf = df.iloc[:,2:]

plt.scatter(xdf,ydf)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(xdf,ydf)
plt.scatter(xdf,ydf)
plt.plot(xdf.values, lr.predict(xdf.values) )
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2)
x_poly = poly_reg.fit_transform(xdf)
x_poly
```

```python
lr.fit(x_poly,ydf)
plt.scatter(xdf,ydf,color = "green")
plt.plot(xdf, lr.predict(x_poly),color="red")
plt.xlabel("Eğitim Seviyesi")
plt.ylabel("Maaş")
plt.title("Polinom Regresyon Tablosu")
```

```python
plt.style.use("ggplot")
poly_reg2 = PolynomialFeatures(degree = 4)
poly_reg3 = PolynomialFeatures(degree = 1)
poly_reg = PolynomialFeatures(degree = 2)
x_poly3 = poly_reg3.fit_transform(xdf)
x_poly2 = poly_reg2.fit_transform(xdf)
x_poly = poly_reg.fit_transform(xdf)
lr.fit(x_poly3,ydf)
plt.plot(xdf, lr.predict(x_poly3),color="purple",label="10.derece",alpha=0.5)
lr.fit(x_poly,ydf)
plt.scatter(xdf,ydf,color = "green")
plt.plot(xdf, lr.predict(x_poly),color="red" ,label="2. derece",alpha=0.5)
lr.fit(x_poly2,ydf)
plt.plot(xdf, lr.predict(x_poly2),color="blue",label="4.derece",alpha=0.5)
print(lr.predict(poly_reg2.fit_transform([[4]])))
plt.xlabel("Eğitim Seviyesi")
plt.ylabel("Maaş")
plt.title("Polinom Regresyon Tablosu")
plt.legend()
```
